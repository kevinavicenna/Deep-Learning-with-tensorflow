```python
#data prep
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
#modeling
from tensorflow.keras.models import Sequential,load_model
from tensorflow.keras.layers import Dense
```

```python
# "/run/media/kevin/DATA/ZENIUS/DATASET/telco_churn.csv"
real_data = pd.read_csv("/run/media/kevin/DATA/PROJECT/Deep-Learning-with-tensorflow/Churn Model/Churn.csv")
```

# Spitting data

```python
data = real_data.copy()
```

```python
data.columns
```

```python
data.columns
```

```python
x = pd.get_dummies(data.drop(['Churn','Customer ID'],axis=1),dtype=int)
y = data['Churn'].apply(lambda x: 1 if x =='Yes' else 0)
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
y
```

```python
x
```

### Build and Compile Model

```python
model = Sequential()
model.add(Dense(units=32,activation='relu',input_dim=len(X_train.columns)))
model.add(Dense(units=64,activation='relu'))
model.add(Dense(units=1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='sgd',metrics='accuracy')
```

```python
y_train
```

```python
X_train
```

### Training & Predict


```python
X_train
```

```python
model.fit(X_train,y_train,epochs=200,batch_size=32)
```

```python
y_pred= model.predict(X_test)
y_pred = [0 if val <0.5 else 1 for val in y_pred]
```

```python
y_pred
```

```python
accuracy_score(y_test,y_pred)
```
